# Feature Engineering

```python
# Splitting dataset
train, val = train_test_split(train_val_data_df_drop, test_size=0.2, random_state=19)

# Check shape of data
print('Train-set Shape : ', train.shape)
print('Validation-set Shape : ', val.shape)
```

```python
# Declare image data generator
trainGen = ImageDataGenerator(
    rotation_range = 45,
    rescale = 1./255,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range = 0.2,
    brightness_range = (0.5, 2),
    validation_split = 0.2 
)

testGen = ImageDataGenerator(
    rescale = 1./255
)
     
```

```python
img_height= 220
img_width= 220
batch_size= 128
```

```python
# Create a pipeline for data
train_ds = trainGen.flow_from_dataframe(dataframe = train,
                                     x_col = 'path',
                                     y_col = 'label',
                                     target_size = (img_height,img_width),
                                     class_mode = 'binary',
                                     batch_size = batch_size,
                                     image_size=(img_height,img_width),
                                     shuffle = True,
                                     subset='training')

val_ds = trainGen.flow_from_dataframe(dataframe = val,
                                   x_col = 'path',
                                   y_col = 'label',
                                   target_size = (img_height,img_width),
                                   class_mode = 'binary',
                                   batch_size = batch_size,
                                   image_size=(img_height,img_width),
                                   shuffle = False,
                                   subset='validation')

test_ds = testGen.flow_from_dataframe(dataframe = test_data_df_drop,
                                     x_col = 'path',
                                     y_col = 'label',
                                     target_size = (img_height,img_width),
                                     class_mode = 'binary',
                                     batch_size = batch_size,
                                     image_size=(img_height,img_width),
                                     shuffle = False)
```

## data augmentation

```python
inputs= (img_height, img_width,3)
data_aug= Sequential(
    [
      layers.RandomFlip('horizontal',
                          input_shape=inputs),
      layers.RandomRotation(0.4),
      layers.RandomZoom(0.2),
    ]
)
```

# Model Training

```python
def plot_accuracy_loss(history):
    fig = plt.figure(figsize=(12,6))

    # Plot accuracy
    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'], label = "acc")
    plt.plot(history.history['val_accuracy'], label = "val_acc")
    plt.title("Training and Validation Accuracy")
    plt.legend()

    # Plot loss function
    plt.subplot(1,2,2)
    plt.plot(history.history['loss'], label = "loss")
    plt.plot(history.history['val_loss'], label = "val_loss")
    plt.title("train_loss vs val_loss")
    plt.legend()

    plt.show()
```

## Model 1

```python
tf.keras.backend.clear_session()
```

```python
# Model definition
model_1 = Sequential([
  data_aug,
  layers.Rescaling(1./255),

  # Convolutional - Pool 1
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(pool_size=(2, 2), padding="valid"),

  # Convolutional - Pool 2
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),

  # Convolutional - Pool 3
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),

  # Convolutional - Pool 4
  layers.Conv2D(128, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),

  # Flatten layer
  layers.Flatten(),

  # Fully connected layer
  layers.Dense(128, activation='relu'),

  # Output layer
  layers.Dense(units=1,activation='sigmoid')
])

# Compile model
model_1.compile(optimizer='adam', loss='binary_crossentropy',
              metrics=['accuracy'])

# Summary model
model_1.summary()
```

```python
epochs= 40 
history_1= model_1.fit(
    train_ds,
    validation_data= val_ds,
    epochs=epochs
)
```

```python
# Save into dataframe
historyBaseDF = pd.DataFrame(history_1.history)
historyBaseDF
```

```python
plot_accuracy_loss(history_1)
```
